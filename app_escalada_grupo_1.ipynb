```python
import sqlite3

conn = sqlite3.connect("database.db")
cursor = conn.cursor()

print("Base de datos creada")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS clientes (
    id INTEGER PRIMARY KEY,
    nombre TEXT,
    correo TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS pedidos (
    id INTEGER PRIMARY KEY,
    cliente_id INTEGER,
    producto TEXT,
    FOREIGN KEY(cliente_id) REFERENCES clientes(id)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS pagos (
    id INTEGER PRIMARY KEY,
    pedido_id INTEGER,
    monto REAL,
    FOREIGN KEY(pedido_id) REFERENCES pedidos(id)
)
""")

print("Tablas creadas")
```

```python
cursor.executemany("INSERT INTO clientes (nombre, correo) VALUES (?, ?)", [
    ("Juli", "juli@mail.com"),
    ("Ana", "ana@mail.com"),
    ("Luis", "luis@mail.com"),
    ("Sofi", "sofi@mail.com"),
    ("Mateo", "mateo@mail.com")
])

cursor.executemany("INSERT INTO pedidos (cliente_id, producto) VALUES (?, ?)", [
    (1, "Hamburguesa"),
    (2, "Pizza"),
    (3, "Sushi"),
    (4, "Tacos"),
    (5, "Pasta")
])

cursor.executemany("INSERT INTO pagos (pedido_id, monto) VALUES (?, ?)", [
    (1, 20000),
    (2, 30000),
    (3, 25000),
    (4, 18000),
    (5, 22000)
])

print("Datos insertados")
```

```python
# READ
cursor.execute("SELECT * FROM clientes")
print("CLIENTES:", cursor.fetchall())

# UPDATE
cursor.execute("UPDATE clientes SET nombre='Maria Juli' WHERE id=1")

# DELETE
cursor.execute("DELETE FROM clientes WHERE id=5")

print("CRUD ejecutado")
```

```python
cursor.execute("""
SELECT clientes.nombre, pedidos.producto
FROM clientes
JOIN pedidos ON clientes.id = pedidos.cliente_id
""")

print("JOIN:", cursor.fetchall())
```

```python
try:
    conn.commit()
    conn.close()
    print("Guardado correctamente")
except Exception as e:
    print("Error:", e)
```

```python
class Usuario:
    def __init__(self, nombre, correo):
        self._nombre = nombre
        self._correo = correo

    def get_nombre(self):
        return self._nombre

    def set_nombre(self, nombre):
        self._nombre = nombre


class Cliente(Usuario):
    def __init__(self, nombre, correo, direccion):
        super().__init__(nombre, correo)
        self.direccion = direccion


class Repartidor(Usuario):
    def __init__(self, nombre, correo, vehiculo):
        super().__init__(nombre, correo)
        self.vehiculo = vehiculo


class Admin(Usuario):
    def __init__(self, nombre, correo):
        super().__init__(nombre, correo)

print("Clases creadas")
```
